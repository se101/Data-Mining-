## 1. Normalization (Standardization)

Let us take the total counts of Anemic Levels in each state so we can analyse state situations from these results, 
The total counts will be ranging from 0 to 2000, and it is easier to apply Normalization techniques here to ease the analysis.

First creating a pandas dataframe of all the counts of Anemic levels in each state, 
The index of the df is the "State and Year" of the counts 
The column headings show the levels of Anemia

```python
main_list = []
index_list = []
values = ['Severe', 'Moderate', 'Mild', 'Not anemic', 'Missing', 'Not applicable']
print(values)

for state in df_sample['V024'].unique():
    #print(df_sample[df_sample['V024']==state]['HW57'].unique())
    list1=[]
    for i in values:
        list1.append(sum(df_sample[df_sample['V024']==state]['HW57']==i))
    #print(list1)
    main_list.append(list1)
    index_list.append(str(state)+' 05-06')
    
    #print(df15[df15['V024']==state]['HW57'].unique())
    list2=[]
    for i in values:
        list2.append(sum(df15[df15['V024']==state]['HW57']==i))
    #print(list2)
    main_list.append(list2)
    index_list.append(str(state)+' 15-16')
    
    #print(df19[df19['V024']==state]['HW57'].unique())
    list3=[]
    for i in values:
        list3.append(sum(df19[df19['V024']==state]['HW57']==i))
    #print(list3)
    main_list.append(list3)
    index_list.append(str(state)+' 19-21')
    
df = pd.DataFrame(main_list, columns = values)
df['State and Year']=index_list
df.set_index('State and Year', inplace=True)
print(df)
```

## 1.1 Min Max Normalization

Min Max Normalization is used to bring the values into a certain range of (min, max) 

```python
df_min_max_scaled = df.copy()
  
# apply normalization techniques
for column in df_min_max_scaled.columns:
    df_min_max_scaled[column] = (df_min_max_scaled[column] - df_min_max_scaled[column].min()) / (df_min_max_scaled[column].max() - df_min_max_scaled[column].min())    
  
# view normalized data
print(df_min_max_scaled)
```

Let us see the plot of the data normalized using Min-Max Normalization

```python
df_min_max_scaled.plot(kind = 'bar')
```

We can now see slight colors of Purple and Blue which were missing in the original plot,

## 1.2 Z-Score Standardization (Normalization)

The distribution is calculated with mean(mu) of 0 and a deviation(sigma) of 1

z_score(x) = ( ( x - mu ) / sigma )

```python
df_z_scaled = df.copy()
  
# apply normalization techniques
for column in df_z_scaled.columns:
    df_z_scaled[column] = (df_z_scaled[column] -
                           df_z_scaled[column].mean()) / df_z_scaled[column].std()    
  
# view normalized data   
display(df_z_scaled)
```

```python
df_z_scaled.plot(kind='bar')
```

##  1.3 Maximum Absolute Scaling

Rescaling of each feature by dividing by Maximum Absolute Value in range -1 to 1

```python
df_max_scaled = df.copy()
  
# apply normalization techniques
for column in df_max_scaled.columns:
    df_max_scaled[column] = df_max_scaled[column]  / df_max_scaled[column].abs().max()
      
# view normalized data
display(df_max_scaled)
```
